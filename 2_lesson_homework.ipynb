Задача 1.
Создать класс Student, который принимает:

class Student_info - объект класса, который хранит информацию о студенте:

имя
возраст
факультет
вид степени (магистратура, бакалавр, докторантура)
scores - список оценок

Требуется:

Добавить метод для Student_info str
Добавить метод average(), возвращающий средний балл студента
Добавить метод passed(), который возвращает True, если средний балл ≥ 70

```python
class Student:
    def __init__(self, name, age, faculty, degree, scores):
        self.name = name
        self.age = age
        self.faculty = faculty
        self.degree = degree
        self.scores = scores

    def student_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Faculty: {self.faculty}, Degree: {self.degree}, Scores: {self.scores}")

    def average(self):
        if len(self.scores) == 0:
            return 0
        return sum(self.scores) / len(self.scores)

    def passed(self):
        return self.average() >= 70



student1 = Student("Nurzhanat", 26, "Economy", "Bachelor", [90, 95, 85])


student1.student_info()
print("Average score", student1.average())
print("Passed", student1.passed())


```

Задача 2.
Создать базовый класс Employee, который хранит:

name
salary
Методы:

info() - возвращает строку с именем и зарплатой Создать два дочерних класса: DataAnalyst - дополнительное поле tools DataScientist — дополнительное поле models Оба дочерних класса должны:
Наследовать от Employee
Переопределять метод info()
Требуется:

Создать несколько объектов разных типов, список сотрудников
Используя цикл, вывести информацию методом info()
Отсортировать всех сотрудников по зарплате
Создать словарь {имя: зарплата}
Вывести только тех, у кого зарплата > 3000

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def info(self):
        return f"{self.name}: {self.salary}$"

emp1 = Employee("Timur", 3100)
emp2 = Employee("Aida", 2000)
emp3 = Employee("Asel", 3500)
emp4 = Employee("Ruslan",2500)
print(emp1.info())
print(emp2.info())
print(emp3.info())
print(emp4.info())

class DataAnalyst(Employee):
     def __init_(self, name, salary, tools):
         super().__init__(name, salary)
         self.tools = tools

def info(self):
        base = super().info()
        return base + f"Tools: {self.tools}"

def info_print(self):
        add_info_worker = super()
        print(add_info_worker)

empdetail = DataAnalyst("Timur", 3100, "Python", "SQL", "PowerBI")
print(empdetail.info())

empdetail.info_print()

```
